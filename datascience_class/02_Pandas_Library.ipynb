```python
# defining function to group hourse power

def hp_cat(x):
  
    if x['hp'] < 100:
        return "Low HP"
  
    elif x['hp'] >= 100 and x['hp'] < 180:
        return "Normal HP"
  
    else:
        return "High HP"
```

```python
cars.apply(hp_cat,axis=1)
```
