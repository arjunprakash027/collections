```python
limit = 100
elon_gray = [None]*limit
j = 0

for i in elon:
    if(j<limit):
        elon_gray[j] = rgb2gray(elon_images[j][:,:,0:3])
        j = j+1
    else:
        break
        
for j in range(100):
    elon =  elon_gray[j]
    elon_gray[j]=resize(elon, (500,500))
    
image_size_elon=elon_gray[6].shape
image_size_elon

flatten_size_elon=image_size_elon[0]*image_size_elon[1]
flatten_size_elon

len_of_images_elon=len(elon_gray)

for i in range(len_of_images_elon):
    elon_gray[i] = np.ndarray.flatten(elon_gray[i]).reshape(flatten_size_elon,1)

elon_gray = np.dstack(elon_gray)
elon_gray = np.rollaxis(elon_gray, axis=2, start=0)
elon_gray.shape
elon_gray = elon_gray.reshape(len_of_images_elon, flatten_size_elon)
elon_gray.shape
elondf = pd.DataFrame(elon_gray)
elondf
elondf["label"] = "elon"
elondf
```
